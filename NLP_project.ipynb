```python
from google.colab import drive
drive.mount('/content/drive')

import os
import nltk
import string
import heapq
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')
nltk.download('stopwords')

folder_path = "/content/drive/MyDrive/BBC News"
output_file_path = "/content/drive/MyDrive/BBC News/BBC_Summary.txt"

categories = ["business", "entertainment"]

with open(output_file_path, "w", encoding="utf-8") as output_file:
    for category in categories:
        output_file.write(f"\n=== ~ Category: {category.upper()} ===\n\n")
        category_path = os.path.join(folder_path, category)

        files = sorted([f for f in os.listdir(category_path) if f.endswith('.txt')])

        # Loop through each article file
        for filename in files:
            file_path = os.path.join(category_path, filename)

            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    article = file.read()

                sentences = sent_tokenize(article)

                stop_words = set(stopwords.words("english"))
                word_freq = {}

                for word in word_tokenize(article.lower()):
                  if word not in stop_words and word not in string.punctuation:
                    word_freq[word] = word_freq.get(word, 0) + 1

                sentence_scores = {}
                for sentence in sentences:
                    words = word_tokenize(sentence.lower())
                    for word in words:
                        if word in word_freq:
                            sentence_scores[sentence] = sentence_scores.get(sentence, 0) + word_freq[word]

                # Get top 2 sentences
                summary_sentences = heapq.nlargest(2, sentence_scores, key=sentence_scores.get)
                summary = " ".join(summary_sentences)

                output_file.write(f"# {filename}\n")
                output_file.write("-> Summary:\n")
                output_file.write(summary + "\n")
                output_file.write("-" * 80 + "\n")

            except Exception as e:
                print(f"Error processing file {filename}: {e}")

print(f"Summaries saved successfully to '{output_file_path}'")
```
